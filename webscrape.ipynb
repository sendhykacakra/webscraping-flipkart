```python
from bs4 import BeautifulSoup 
import requests 
import csv
import pandas as pd
```

```python
req = requests.get("https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1") 
content = req.content
```

```python
soup = BeautifulSoup(content,'html.parser')
print(soup.prettify())
```

```python
desc = soup.find_all('div' , class_='_4rR01T')
desc
```

```python
# Create a list to store the descriptions
descriptions = [] 
for i in range(len(desc)):
    descriptions.append(desc[i].text)
len(descriptions)
```
